```python
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
import time
import random
```

**OUTDOOR LIVING**

```python
# adding class/catgeory of our images, ensure that they are separated by correct folder names
DATADIR = "/content/drive/My Drive/Colab Notebooks/Capstone/testing/train_images/Outdoor_Living"
CATEGORIES = ['BACKYARD TRAMPOLINES AND ACCESSORIES',
              'BBQ COVERS',
              'BBQ TRENDS 2021',
              'BEVERAGE CARTS TUBS AND COOLERS',
              'CHAINSAWS',
              'CHARCOAL BBQS',
              'DECK & FENCE LIGHTING',
              'DECK BOXES & STORAGE BENCHES',
              'ELECTRIC BBQS',
              'FIRE BOWLS',
              'FIRE TABLES',
              'FIREPITS',
              'FIREWOOD STORAGE',
              'GARBAGE CANS & BAGS',
              'GARDEN HOSES',
              'GAS BBQS',
              'GAZEBOS PERGOLAS AND CANOPIES',
              'GENERATORS AND ACCESSORIES',
              'GOGGLES SNORKELS AND SWIM CAPS',
              'GRASS TRIMMERS',
              'GREENHOUSES',
              'HAMMOCKS AND PATIO SWINGS',
              'HARD COOLERS',
              'HEDGE TRIMMERS',
              'HOSE NOZZLES AND WANDS',
              'INFLATABLE BOUNCERS',
              'LAWN MOWERS',
              'LOG SPLITTERS, WOOD CHIPPERS AND SHREDDERS',
              'OUTDOOR COOKERS AND FRYERS',
              'OUTDOOR HEATING',
              'PATIO CUSHIONS AND SEAT PADS',
              'POOLS',
              'POTS AND PLANTERS',
              'SKATEBOARDS AND SCOOTERS',
              'SNOWBLOWERS',
              'SOFT COOLERS',
              'WATER TIMERS',
              'WATERING CANS']

IMG_SIZE = 256
training_data = []

start = time.time()
def create_training_data():
    for category in CATEGORIES:
        print('--------{}---------'.format(category))
        path = os.path.join(DATADIR, category) # path to category images
        class_num = CATEGORIES.index(category)
        for img in os.listdir(path):
            try:
                img_array = cv2.imread(os.path.join(path,img), cv2.IMREAD_GRAYSCALE)
                new_array = cv2.resize(img_array, (IMG_SIZE, IMG_SIZE))
                training_data.append([new_array, class_num])
            except Exception as e:
                pass

create_training_data()
random.shuffle(training_data)
end = time.time()
print(end-start)
```

```python
# Labeling the training data
X = [] #features
y = [] #label

for features,label in training_data:
    X.append(features)
    y.append(label)

X = np.array(X).reshape(-1, IMG_SIZE, IMG_SIZE, 1)
y = np.array(y)
```

```python
# Saving the X and y values to pickle for future use
# This is important when handling several datasets to prevent from labeling repeatedly

import pickle

pickle_out = open("/content/drive/My Drive/Colab Notebooks/Capstone/testing/X_Outdoor_Living.pickle", "wb")
pickle.dump(X, pickle_out)
pickle_out.close

pickle_out = open("/content/drive/My Drive/Colab Notebooks/Capstone/testing/y_Outdoor_Living.pickle", "wb")
pickle.dump(y, pickle_out)
# pickle_out.close
```
